```python
# Loaded variable 'df' from URI: /Users/nikolya/code/lab/BiopolymersKG/data/AptaBench_dataset_v2.csv
import pandas as pd
df = pd.read_csv(r'/Users/nikolya/code/lab/BiopolymersKG/data/AptaBench_dataset_v2.csv')
```

```python
groups = df.groupby('type')
```

```python
df1, df2 = groups
```

```python
df_dna = df1[1]
df_rna = df2[1]
```

```python
df_dna
```

```python
import numpy as np

uniq_dna = df_dna[['sequence', 'source']].drop_duplicates()
uniq_dna = uniq_dna.reset_index(drop=True)
uniq_dna['name'] = ['dna_'] * uniq_dna.shape[0]
uniq_dna['name'] = uniq_dna['name'] + uniq_dna['source'] + (['_'] * uniq_dna.shape[0]) + uniq_dna.index.astype(str)

uniq_dna
```

```python
uniq_rna = df_rna[['sequence', 'source']].drop_duplicates()
uniq_rna = uniq_rna.reset_index(drop=True)
uniq_rna['name'] = ['rna_'] * uniq_rna.shape[0]
uniq_rna['name'] = uniq_rna['name'] + uniq_rna['source'] + (['_'] * uniq_rna.shape[0]) + uniq_rna.index.astype(str)

uniq_rna
```

```python
uniq_small_molecule = df[['canonical_smiles', 'source']].drop_duplicates()
uniq_small_molecule = uniq_small_molecule.reset_index(drop=True)
uniq_small_molecule['name'] = ['small_molecule_'] * uniq_small_molecule.shape[0]
uniq_small_molecule['name'] = uniq_small_molecule['name'] + uniq_small_molecule['source'] + (['_'] * uniq_small_molecule.shape[0]) + uniq_small_molecule.index.astype(str)

uniq_small_molecule
```

```python
# Переименовываем колонку name в aptamer_name для DNA и RNA
uniq_dna_renamed = uniq_dna.rename(columns={'name': 'aptamer_name'})
uniq_rna_renamed = uniq_rna.rename(columns={'name': 'aptamer_name'})

# Переименовываем колонку name в small_molecule_name для малых молекул
uniq_small_molecule_renamed = uniq_small_molecule.rename(columns={'name': 'small_molecule_name'})
```

```python
# Мерджим DNA аптамеры
df_merged = df.merge(uniq_dna_renamed, on=['sequence', 'source'], how='left')

# Мерджим RNA аптамеры (объединяем с той же колонкой aptamer_name)
# Сначала заполняем aptamer_name из RNA там, где она пустая
df_merged = df_merged.merge(uniq_rna_renamed, on=['sequence', 'source'], how='left', suffixes=('', '_rna'))

# Объединяем колонки aptamer_name и aptamer_name_rna в одну
df_merged['aptamer_name'] = df_merged['aptamer_name'].fillna(df_merged['aptamer_name_rna'])
df_merged = df_merged.drop(columns=['aptamer_name_rna'])

# Мерджим малые молекулы
df_merged = df_merged.merge(uniq_small_molecule_renamed, on=['canonical_smiles', 'source'], how='left')

df_merged
```

```python
# Проверяем результат
print(f"Исходный размер: {df.shape}")
print(f"Размер после мерджа: {df_merged.shape}")
print(f"\nКолонки: {df_merged.columns.tolist()}")
print(f"\nПроверка заполненности aptamer_name: {df_merged['aptamer_name'].notna().sum()} из {len(df_merged)}")
print(f"Проверка заполненности small_molecule_name: {df_merged['small_molecule_name'].notna().sum()} из {len(df_merged)}")

# Показываем несколько первых строк
df_merged[['type', 'sequence', 'canonical_smiles', 'aptamer_name', 'small_molecule_name', 'source']].head(10)
```

```python
# Создаем датафрейм для DNA аптамеров
dna_final = df_merged[df_merged['type'] == 'DNA'][['aptamer_name', 'sequence']].copy()
dna_final = dna_final.rename(columns={'aptamer_name': 'name', 'sequence': 'content'})
dna_final = dna_final.drop_duplicates().reset_index(drop=True)

print(f"DNA аптамеры: {dna_final.shape[0]} уникальных записей")
dna_final.head()
```

```python
# Создаем датафрейм для RNA аптамеров
rna_final = df_merged[df_merged['type'] == 'RNA'][['aptamer_name', 'sequence']].copy()
rna_final = rna_final.rename(columns={'aptamer_name': 'name', 'sequence': 'content'})
rna_final = rna_final.drop_duplicates().reset_index(drop=True)

print(f"RNA аптамеры: {rna_final.shape[0]} уникальных записей")
rna_final.head()
```

```python
# Создаем датафрейм для малых молекул
small_molecules_final = df_merged[['small_molecule_name', 'canonical_smiles']].copy()
small_molecules_final = small_molecules_final.rename(columns={'small_molecule_name': 'name', 'canonical_smiles': 'content'})
small_molecules_final = small_molecules_final.drop_duplicates().reset_index(drop=True)

print(f"Малые молекулы: {small_molecules_final.shape[0]} уникальных записей")
small_molecules_final.head()
```

```python
# Итоговая статистика
print("="*60)
print("ИТОГОВАЯ СТАТИСТИКА")
print("="*60)
print(f"DNA аптамеры:      {dna_final.shape[0]:5} уникальных записей")
print(f"RNA аптамеры:      {rna_final.shape[0]:5} уникальных записей")
print(f"Малые молекулы:    {small_molecules_final.shape[0]:5} уникальных записей")
print("="*60)
print(f"Всего записей:     {dna_final.shape[0] + rna_final.shape[0] + small_molecules_final.shape[0]:5}")
print("="*60)
```

```python
dna_final.to_csv('../data/aptabench_dna_final.csv')
rna_final.to_csv('../data/aptabench_rna_final.csv')
small_molecules_final.to_csv('../data/aptabench_small_molecules_final.csv')
df_merged.to_csv('../data/aptabench_df_merged.csv')
```

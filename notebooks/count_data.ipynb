```python
def count_samples(
        path: str,
        dataset_name: str,
        entity_fname_1: str,
        entity_fname_2: str,
        relationship_fname: str,
        ) -> pd.DataFrame:
    """
    Works only with csv files
    
    returns: entities count, relationships count
    """

    entity_ds_1 = pd.read_csv(path + entity_fname_1)
    if 'subtype' in entity_ds_1.columns:
        g1 = entity_ds_1.groupby(['type', 'subtype']).count()[['name']]
    else:
        g1 = entity_ds_1.groupby(['type']).count()[['name']]


    entity_ds_2 = pd.read_csv(path + entity_fname_2)
    if 'subtype' in entity_ds_2.columns:
        g2 = entity_ds_2.groupby(['type', 'subtype']).count()[['name']]
    else:
        g2 = entity_ds_2.groupby(['type']).count()[['name']]


    relationship_ds = pd.read_csv(path + relationship_fname)
    

    entities_count = entity_ds_1.shape[0] + entity_ds_2.shape[0]
    relationships_count = relationship_ds.shape[0]

    return dataset_name, entities_count, relationships_count
```

### сравнение лейблов в датасетах и в базе данных

```python
labels = pd.read_excel('./labels_comparison.xlsx')
labels
```

dataset_label - сочетние тип-подтип в датасетах  
database_label - это же сочетание, но в базе данных

видно, что лейблы в датасетах по формату отличаются от лейблов в базе данных, поэтому нужно их приобразовать к одному виду, чтобы можно было сравнивать

```python
def transform_ds_label(x: str) -> str:
    return x.replace('(', '').replace(')', '').replace("'", "").replace(',', ':').replace(' ', '')

def transform_db_label(x: str) -> str:
    return x.replace('[', '').replace(']', '').replace(',', ':').replace(' ', '')
```

```python
labels['dataset_label'] = labels['dataset_label'].apply(transform_ds_label)
labels['database_label'] = labels['database_label'].apply(transform_db_label)
```

```python
labels
```

также можно заметить, что лейблы в датасете и в базе данных могут идти в разных последовательностях
чтобы это нивелировать, введем колонку, которая не зависит от порядка следования элементов

```python
labels['dataset_label'] = labels['dataset_label'].apply(lambda x: ':'.join(set(x.split(':'))))
labels['database_label'] = labels['database_label'].apply(lambda x: ':'.join(set(x.split(':'))))
```

### сравнение связей в датасетах и базе данных

```python
rels = pd.read_excel('./relationships_comparison.xlsx')
rels
```

обработаем лейблы

```python
def transform_ds_label(x: str) -> str:
    x = sorted(x.replace('(', '').replace(')', '').replace("'", "").split(', '))
    if len(x) == 1:
        return x[0]
    elif len(x) == 3:
        first = x[:2]
        last = x[-1]
        return ':'.join(first) + ' || ' + last
    else:
        first = x[:len(x) // 2]
        last = x[len(x) // 2:]
        return ':'.join(first) + ' || ' + ':'.join(last)


def transform_db_label(x: str) -> str:
    return x.replace('[', '').replace(']', '').replace(',', ':').replace(' ', '')
```

```python
rels['dataset_labels'] = rels['dataset_labels'].apply(transform_ds_label)
```

```python
rels['database_label_1'] = rels['database_label_1'].apply(transform_db_label)
rels['database_label_2'] = rels['database_label_2'].apply(transform_db_label)
rels['database_labels'] = rels['database_label_1'] + ' || ' + rels['database_label_2']

clean_rels = rels.drop(['database_label_1', 'database_label_2'], axis=1)

clean_rels
```

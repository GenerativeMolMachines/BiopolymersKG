```python
import pandas as pd
import numpy as np
```

```python
def prepare_dataset(ds: pd.DataFrame):
    ds = ds.drop_duplicates()
    if 'type' in ds.columns:
        ds['type'] = ds['type'].str.lower().str.strip()

    if 'subtype' in ds.columns:
        if ds['subtype'].isna().sum() == ds.shape[0]:
            ds.drop('subtype', axis=1, inplace=True)
        else:
            ds['subtype'] = ds['subtype'].str.lower().str.strip()

    if 'name' in ds.columns:
        ds.dropna(subset='name', inplace=True)

    ds.replace(np.nan, None, inplace=True)
    return ds
```

### binding db

```python
protein_data = pd.read_csv("binding_db/protein_data.csv")
protein_data
```

```python
protein_data['subtype'].unique()
```

```python
protein_data = prepare_dataset(protein_data)
protein_data
```

```python
protein_data.to_csv("binding_db/prepared/protein_data.csv", index=False)
```
